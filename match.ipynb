```python
import pandas as pd
import numpy as np
```

```python
accu_test_df = pd.read_csv('batch_outputs/accu_test_scored.csv')
accu_train_df = pd.read_csv('batch_outputs/accu_train_scored.csv')
redm_test_df = pd.read_csv('batch_outputs/redm_test_scored.csv')
redm_train_df = pd.read_csv('batch_outputs/redm_train_scored.csv')
```

```python
for name, df in [
    ("accu_test_df", accu_test_df),
    ("accu_train_df", accu_train_df),
    ("redm_test_df", redm_test_df),
    ("redm_train_df", redm_train_df),
]:
    pct = (df["actual_is_anomaly"] == df["predicted_is_anomaly"]).mean() * 100
    print(f"{name}: {pct:.2f}%")
```
